```python
import os
import pandas as pd
```

```python
comment_senti_result_path = r"\專案實作\地區餐廳評論分析\google_restaurant_comments_txt\restaurant_comments_processed"

for root, dirs, files in os.walk(comment_senti_result_path):
    for dir in dirs:
        print(dir)
    for file_name in files:
        # print(os.path.join(root,file_name))
        
        if file_name.endswith("_result.xlsx"):
            excel_file_path = os.path.join(root,file_name)
            file_name = os.path.basename(excel_file_path)
            
            df_senti = pd.read_excel(excel_file_path)
            
            restaurant_name = excel_file_path.split('.')[0].replace(root,'').replace('\Comment_of_','').replace('_result','')
            total_comments_count = len(df_senti)
            positive_comments_count = len(df_senti[df_senti['情感分析結果']=='正面'])
            conflict_comments_count = len(df_senti[df_senti['情感分析結果']=='衝突'])
            negative_comments_count = len(df_senti[df_senti['情感分析結果']=='負面'])
            neutral_comments_count = len(df_senti[df_senti['情感分析結果']=='中性'])
            
            df_senti_statistic_result = pd.DataFrame({
                'RESTAURANT':[restaurant_name],
                'FETCH_COMMENTS_COUNT':[total_comments_count],
                'POSITIVE_COMMENTS_COUNT':[positive_comments_count],
                'CONFLICT_COMMENTS_COUNT':[conflict_comments_count],
                'NEGATIVE_COMMENTS_COUNT':[negative_comments_count],
                'NEUTRAL_COMMENTS_COUNT':[neutral_comments_count]
            })
            
            df_senti_statistic_result['positive_percentage'] = (df_senti_statistic_result['POSITIVE_COMMENTS_COUNT'] / df_senti_statistic_result['FETCH_COMMENTS_COUNT']).round(2)
            df_senti_statistic_result['conflict_percentage'] = (df_senti_statistic_result['CONFLICT_COMMENTS_COUNT'] / df_senti_statistic_result['FETCH_COMMENTS_COUNT']).round(2)
            df_senti_statistic_result['negative_percentage'] = (df_senti_statistic_result['NEGATIVE_COMMENTS_COUNT'] / df_senti_statistic_result['FETCH_COMMENTS_COUNT']).round(2)
            df_senti_statistic_result['neutral_percentage'] = (df_senti_statistic_result['NEUTRAL_COMMENTS_COUNT'] / df_senti_statistic_result['FETCH_COMMENTS_COUNT']).round(2)
            
            senti_result_stat_path = excel_file_path.replace('_result.xlsx','_result_statistic.xlsx')
            # print(df_senti_statistic_result)
            
            # print(senti_result_stat_path)
            df_senti_statistic_result.to_excel(senti_result_stat_path,index=False)
```
