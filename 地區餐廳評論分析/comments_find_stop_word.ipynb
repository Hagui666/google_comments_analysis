```python
import jieba

# 打开评论文件
with open(r"\專案實作\地區餐廳評論分析\google_restaurant_comments_txt\restaurant_comments_unprocessed\Buffet_restaurant_comments_txt\Comment_of_INPARADISE 饗饗 信義店.txt", 'r', encoding='utf-8') as file:
    comments = file.read().replace('[','').replace(']','').replace('\'','')

# 分词
seg_list = jieba.cut(comments, cut_all=False)

# 打开停用词文件
with open(r"\demo\專案實作\地區餐廳評論分析\stop_word.txt", 'r', encoding='utf-8') as stop_word_file:
    stopwords = set(stop_word_file.read().splitlines())

# 去除停用词并重新组合评论
filtered_comments = []
for word in seg_list:
    if word not in stopwords and word.strip() != '':
        filtered_comments.append(word)

# 打印处理后的评论
# for i, word in enumerate(filtered_comments):
    # print(f'处理后的词汇{i + 1}: {word}')

# print(filtered_comments)    
    
```
