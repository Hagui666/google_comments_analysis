```python
import jieba.analyse
import jieba
import os
from collections import Counter
```

```python
comment_unprocessed_txt_path = r"\專案實作\地區餐廳評論分析\google_restaurant_comments_txt\restaurant_comments_unprocessed"

top_50_key_words = []

for root, dirs, files in os.walk(comment_unprocessed_txt_path):
    # for dir in dirs:
        # print(dir)
    for file_name in files:
        # print(os.path.join(root,file_name))
        
        if file_name.endswith(".txt"):
            txt_file_path = os.path.join(root,file_name)
            file_name = os.path.basename(txt_file_path)
            
            with open(txt_file_path,'r',encoding='utf-8') as file:
                content = file.readline()
                print(file)
            key_words = jieba.analyse.extract_tags(content,topK=50)
            # print(key_words)
            top_50_key_words.extend(key_words)
```

```python
word_count = Counter(top_50_key_words)
top_50_key_words = sorted(top_50_key_words,key=lambda x:word_count[x],reverse=True)
top_50_key_words
```

```python
top_50_key_words = list(set(top_50_key_words))
top_50_key_words
```

```python
stop_word_txt_path = r"\專案實作\地區餐廳評論分析\stop_word(demo).txt"

with open(stop_word_txt_path,'w') as file:
    list_content = ','.join(top_50_key_words)
    file.write(list_content)
```
